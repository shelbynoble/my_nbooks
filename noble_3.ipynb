# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide.plot.line(y=['15-19 Years'], title='Child Mortality: AgeGroup 15-19 Years', legend=False)
```

```python
mortality_wide.plot.area(legend='reverse')
```

```python
mortality_wide.query('Year in [2000]').plot.bar(title='Child Mortality: Year 2000', xlabel= ' ')
```

```python
mortality_data.query('Year in [2000]').pivot(index='Year', columns='AgeGroup', values='DeathRate').plot.bar(title='Child Mortality: Year 2000', xlabel= ' ')
```

```python
mortality_data.plot.hist(y='DeathRate', bins=15)
```

```python
mortality_data.query('Year in [1900]').plot.density(y='DeathRate', title= 'Distribution of Death Rates: Year 1900', grid=True)
```

```python
mortality_data.query('Year in [1900, 1925, 1950, 1975, 2000]').groupby('Year')['DeathRate'].sum().plot.pie()
```

```python
 mortality_wide.query('Year in [1900, 1925, 1950, 1975, 2000]').plot.barh(
     title=['Child Mortality: Age 01-04', 'Child Mortality: Age 05-09',
            'Child Mortality: Age 10-14', 'Child Mortality: Age 15-19'],
     subplots=True, layout=(2,2), sharey=True, legend=False, grid=False, xlabel='Deaths Per 100,000')
```
